```python
def get_lndetDdw(Ls_, seed_, width_):
    tmp = np.loadtxt("ev_Ddw_Ls"+str(Ls_)+"_seed"+str(seed_)+"_width"+str(width_)+".dat")
    eig = np.array([tmp.T[1], tmp.T[2]]).T
    eig_cplx = eig.T[0] + 1.0j * eig.T[1]
    
    return np.sum( np.log(eig_cplx) ).real
```

```python
Ls_s=[4, 6, 8, 10, 12]
```
